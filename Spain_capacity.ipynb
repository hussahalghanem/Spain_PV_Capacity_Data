```python
def to_df(cap):
    # extract the included data
    included_data = cap["included"]
    # create a dataframe from the included data
    df = pd.json_normalize(
        included_data, 
        record_path=["attributes", "values"], 
        meta=["id", ["attributes", "title"], ["attributes", "color"], ["attributes", "type"], ["attributes", "composite"], ["attributes", "last-update"]], 
        errors='ignore'
    )
    # select solar PV data only
    df = df[df['attributes.title'] == 'Solar photovoltaic']
    return df
```

```python
# convert data to df format
Andalucia_2015_2018 = to_df(Andalucia_2015_2018)
Andalucia_2019_2023 = to_df(Andalucia_2019_2023)

Aragon_2015_2018 = to_df(Aragon_2015_2018)
Aragon_2019_2023 = to_df(Aragon_2019_2023)

Cantabria_2015_2018 = to_df(Cantabria_2015_2018)
Cantabria_2019_2023 = to_df(Cantabria_2019_2023)

Castilla_la_Mancha_2015_2018 = to_df(Castilla_la_Mancha_2015_2018)
Castilla_la_Mancha_2019_2023 = to_df(Castilla_la_Mancha_2019_2023)

Castilla_y_Leon_2015_2018 = to_df(Castilla_y_Leon_2015_2018)
Castilla_y_Leon_2019_2023 = to_df(Castilla_y_Leon_2019_2023)

Cataluna_2015_2018 = to_df(Cataluna_2015_2018)
Cataluna_2019_2023 = to_df(Cataluna_2019_2023)

Comunidad_de_Navarra_2015_2018 = to_df(Comunidad_de_Navarra_2015_2018)
Comunidad_de_Navarra_2019_2023 = to_df(Comunidad_de_Navarra_2019_2023)

Comunidad_de_Madrid_2015_2018 = to_df(Comunidad_de_Madrid_2015_2018)
Comunidad_de_Madrid_2019_2023 = to_df(Comunidad_de_Madrid_2019_2023)

Extremadura_2015_2018 = to_df(Extremadura_2015_2018)
Extremadura_2019_2023 = to_df(Extremadura_2019_2023)

Galicia_2015_2018 = to_df(Galicia_2015_2018)
Galicia_2019_2023 = to_df(Galicia_2019_2023)

La_Rioja_2015_2018 = to_df(La_Rioja_2015_2018)
La_Rioja_2019_2023 = to_df(La_Rioja_2019_2023)

Pais_Vasco_2015_2018 = to_df(Pais_Vasco_2015_2018)
Pais_Vasco_2019_2023 = to_df(Pais_Vasco_2019_2023)

Principado_de_Asturias_2015_2018 = to_df(Principado_de_Asturias_2015_2018)
Principado_de_Asturias_2019_2023 = to_df(Principado_de_Asturias_2019_2023)

Region_de_Murcia_2015_2018 = to_df(Region_de_Murcia_2015_2018)
Region_de_Murcia_2019_2023 = to_df(Region_de_Murcia_2019_2023)

Peninsular_2015_2018 = to_df(Peninsular_2015_2018)
Peninsular_2019_2023 = to_df(Peninsular_2019_2023)

Spain_2015_2018 = to_df(Spain_2015_2018)
Spain_2019_2023 = to_df(Spain_2019_2023)
```

```python
# concatenate dataframes for each region and assign region name
Andalucia = pd.concat([Andalucia_2015_2018, Andalucia_2019_2023], axis=0).assign(Region='Andalucia', NUTS2='ES61')

Aragon = pd.concat([Aragon_2015_2018, Aragon_2019_2023], axis=0).assign(Region='Aragon', NUTS2='ES24')

Cantabria = pd.concat([Cantabria_2015_2018, Cantabria_2019_2023], axis=0).assign(Region='Cantabria', NUTS2='ES13')

Castilla_la_Mancha = pd.concat([Castilla_la_Mancha_2015_2018, Castilla_la_Mancha_2019_2023], axis=0).assign(Region='Castilla_la_Mancha', NUTS2='ES42')

Castilla_y_Leon = pd.concat([Castilla_y_Leon_2015_2018, Castilla_y_Leon_2019_2023], axis=0).assign(Region='Castilla y Leon', NUTS2='ES41')

Cataluna = pd.concat([Cataluna_2015_2018, Cataluna_2019_2023], axis=0).assign(Region='Cataluna', NUTS2='ES51')

Comunidad_de_Navarra = pd.concat([Comunidad_de_Navarra_2015_2018, Comunidad_de_Navarra_2019_2023], axis=0).assign(Region='Comunidad de Navarra', NUTS2='ES22')

Comunidad_de_Madrid = pd.concat([Comunidad_de_Madrid_2015_2018, Comunidad_de_Madrid_2019_2023], axis=0).assign(Region='Comunidad de Madrid', NUTS2='ES30')

Extremadura = pd.concat([Extremadura_2015_2018, Extremadura_2019_2023], axis=0).assign(Region='Extremadura', NUTS2='ES43')

Galicia = pd.concat([Galicia_2015_2018, Galicia_2019_2023], axis=0).assign(Region='Galicia', NUTS2='ES11')

La_Rioja = pd.concat([La_Rioja_2015_2018, La_Rioja_2019_2023], axis=0).assign(Region='La Rioja', NUTS2='ES23')

Pais_Vasco = pd.concat([Pais_Vasco_2015_2018, Pais_Vasco_2019_2023], axis=0).assign(Region='Pais Vasco', NUTS2='ES21')

Principado_de_Asturias = pd.concat([Principado_de_Asturias_2015_2018, Principado_de_Asturias_2019_2023], axis=0).assign(Region='Principado de Asturias', NUTS2='ES12')

Region_de_Murcia = pd.concat([Region_de_Murcia_2015_2018, Region_de_Murcia_2019_2023], axis=0).assign(Region='Region de Murcia', NUTS2='ES62')

Peninsular = pd.concat([Peninsular_2015_2018, Peninsular_2019_2023], axis=0).assign(Region='Peninsular')

Spain = pd.concat([Spain_2015_2018, Spain_2019_2023], axis=0).assign(Region='Spain')
```
